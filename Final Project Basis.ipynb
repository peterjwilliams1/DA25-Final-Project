```python
from geopy.geocoders import Nominatim
import pandas as pd
import time
import logging

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger()

# Initialize geolocator
geolocator = Nominatim(user_agent="petiiii")

# Create an empty list to store addresses
addresses = []

# Loop through each row in the DataFrame
for index, row in df_filtered_years.iterrows():
    latitude = row['lat']
    longitude = row['long']
    coordinates = (latitude, longitude)
    try:
        # Perform reverse geocoding
        location = geolocator.reverse(coordinates, timeout=10)
        address = location.address if location else None
    except Exception as e:
        logger.error(f"Error for {coordinates}: {e}")
        address = None
    addresses.append(address)
    # Sleep for a second to respect rate limitations
    time.sleep(1)

# Create a new DataFrame including the 'address' column
df_filtered_years_with_addresses = df_filtered_years.copy()
df_filtered_years_with_addresses['address'] = addresses

# Save DataFrame to CSV
output_file = "geocoded_addresses.csv"
df_filtered_years_with_addresses.to_csv(output_file, index=False)

logger.info(f"CSV file saved as {output_file}")
logger.info(df_filtered_years_with_addresses.head())
```

## Pearson Correlation: Does Meteorite Mass Depend on Year?

Pearson Correlation: Does Meteorite Mass Depend on Year?

Steps:
Define Hypothesis:
Null Hypothesis (𝐻0): No correlation between year and mass.
Alternative Hypothesis (𝐻𝑎): Mass is correlated with year.

Use Pearson’s correlation test to measure the relationship.
Interpret the correlation coefficient (r):

r=1 → Perfect positive correlation (mass increases with year).
r=−1 → Perfect negative correlation (mass decreases with year).
r=0 → No correlation.

## Chi-Square Test: Are Meteorite Falls Evenly Distributed Across Decades?

Chi-Square Test: Are Meteorite Falls Evenly Distributed Across Decades?
We can check if meteorites fell uniformly across decades or if some decades had significantly more falls than others.

Steps:
Define Hypothesis:

Null Hypothesis (𝐻0): Meteorites fall evenly across decades.
Alternative Hypothesis (𝐻𝑎): Meteorites do not fall evenly.

Expected vs. Observed Frequencies:

If meteorites fell randomly, each decade should have a similar number.
Compare observed data with expected counts using a Chi-Square Test.

Explanation:
Chi-Square Test compares observed (actual falls per decade) vs. expected (if falls were evenly spread).
p-value < 0.05 → Significant difference, meaning meteorites don’t fall evenly.
p-value > 0.05 → No significant difference, meaning falls are random.

```python
import pandas as pd
import scipy.stats as stats

# Sample dataset (replace with full dataset)
data = {
    "decade": [1880, 1950, 1950, 1970, 1900, 1950, 2000, 1880, 1900, 1970]
}

df = pd.DataFrame(data)

# Step 1: Count meteorites per decade
observed_counts = df["decade"].value_counts()

# Step 2: Compute expected counts (equal distribution assumption)
expected_counts = [len(df) / len(observed_counts)] * len(observed_counts)

# Step 3: Perform Chi-Square Test
chi2_stat, p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Step 4: Interpret results
print(f"Chi-Square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: Meteorite falls are NOT evenly distributed across decades.")
else:
    print("Fail to reject the null hypothesis: Meteorite falls appear evenly distributed.")
```

## Kolmogorov-Smirnov (KS) Test: Are Meteorite Masses Normally Distributed?

Kolmogorov-Smirnov (KS) Test: Are Meteorite Masses Normally Distributed?
This checks if meteorite masses follow a normal distribution.

Steps:
Null Hypothesis (𝐻0): Data is normally distributed.
Alternative Hypothesis (𝐻𝑎): Data is not normally distributed.

KS Test compares actual distribution vs. normal distribution.

Explanation:
p-value < 0.05 → Data not normally distributed (use non-parametric tests like Mann-Whitney).
p-value > 0.05 → Data follows normal distribution (use t-tests, ANOVA).

```python
from scipy.stats import kstest

# Step 1: Perform KS test for normality
ks_stat, p_value = kstest(df["mass"], "norm", args=(df["mass"].mean(), df["mass"].std()))

# Step 2: Interpret results
print(f"KS Statistic: {ks_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: Meteorite masses are NOT normally distributed.")
else:
    print("Fail to reject the null hypothesis: Meteorite masses follow a normal distribution.")
```

## Predicting Meteorite Impact Mass (Regression)

Train a regression model (Linear Regression, Random Forest) to predict meteorite mass based on:

Year (when it fell)
Latitude & Longitude (where it fell)
Recclass (meteorite type)

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Load dataset
df = pd.read_csv("meteorite_data.csv")  # Replace with actual dataset

# Select relevant columns
df = df[["year", "lat", "long", "recclass", "mass"]].dropna()

# Encode categorical variable (recclass)
le = LabelEncoder()
df["recclass_encoded"] = le.fit_transform(df["recclass"])

# Define features (X) and target (y)
X = df[["year", "lat", "long", "recclass_encoded"]]
y = df["mass"]

# Scale features (optional, for better model performance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training & test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### Train Regression Models

#### Random Forest Regressor

```python
# Train a Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf.predict(X_test)

# Evaluate performance
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"Random Forest MAE: {mae_rf:.2f}")
print(f"Random Forest R² Score: {r2_rf:.2f}")
```

Step 2: Train the data

```python
from sklearn.linear_model import LinearRegression

# Train a Linear Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Make predictions
y_pred_lr = lr.predict(X_test)

# Evaluate performance
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print(f"Linear Regression MAE: {mae_lr:.2f}")
print(f"Linear Regression R² Score: {r2_lr:.2f}")
```

Step 3: Predict Mass for a New Meteorite

```python
new_meteorite = np.array([[2025, 35.0, -90.0, le.transform(["H5"])[0]]])  # Example new impact
new_meteorite_scaled = scaler.transform(new_meteorite)

predicted_mass = rf.predict(new_meteorite_scaled)
print(f"Predicted Meteorite Mass: {predicted_mass[0]:.2f} grams")
```

Key Takeaways:
Random Forest performs better than Linear Regression due to its ability to capture non-linear patterns.
The MAE (Mean Absolute Error) shows how far our predictions are from actual masses.
You can experiment with different models (Gradient Boosting, XGBoost) for better accuracy.

## Predicting Meteorite Fall Likelihood (Classification)

Train a classification model (Logistic Regression, Decision Trees) to predict whether a meteorite will fall in a given location.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Convert 'fall' column to binary (1 = Fell, 0 = Not Fell)
df["fall_binary"] = df["fall"].apply(lambda x: 1 if x == "Fell" else 0)

# Define features (X) and target (y)
X = df[["lat", "long", "year", "recclass_encoded"]]
y = df["fall_binary"]

# Scale features
X_scaled = scaler.fit_transform(X)

# Split into training & test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

#### Train Classification Models

#### Random Forest Classifier

```python
# Train a Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
print("Classification Report:\n", classification_report(y_test, y_pred))
```

#### Logistic Regression (Baseline Model)

```python
from sklearn.linear_model import LogisticRegression

# Train a Logistic Regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Make predictions
y_pred_logreg = logreg.predict(X_test)

# Evaluate performance
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print(f"Logistic Regression Accuracy: {accuracy_logreg:.2f}")
```
